## **3. Regression**
</div>



<font color='darkred'>**Regression analysis**</font>  is a statistical technique for analysing and comprehending the connection between two or more variables of interest. The methodology used to do regression analysis aids in understanding which elements are significant, which may be ignored, and how they interact with one another.

In this part, we will use the module called `sklearn` and `statsmodels` to make regression analysis and its report. 


**Installation**

*   Anaconda: `conda install -c conda-forge statsmodels`
*   Pip: `!pip install statsmodels` and `!pip install scikit-learn`


Check out more: [statsmodels documentation](https://www.statsmodels.org/stable/examples/index.html) and below is the example of regression in our daily life of working that we probably meet. 

validation_plot_Meedhoo_West.png

### Import module

```python
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
```

### Load dataset from folder

```python
# import data
ds = pd.read_csv('data/data.csv')

print(len(ds))
ds.head()
```

```python
# Observation 
ds_obs = ds['obs']

ds_obs
```

```python
# Visualize data
ds_obs.plot()
```

```python
# Simulation 
ds_sim = ds['sim']

ds_sim
```

```python
# Visualize data
ds_sim.plot()
```

### Calculate metrics

```python
# Calculate R-squared
R2 = r2_score(ds_obs, ds_sim)

print("R squared: %.3f" % R2)
```

```python
# Calculate MSE
MSE = mean_squared_error(ds_obs, ds_sim, squared=False)

print("Root Mean Squared Error: %.3f" % MSE)
```

```python
# Calculate RMSE
RMSE = mean_squared_error(ds_obs, ds_sim, squared=True)

print("Mean Squared Error: %.3f" % RMSE)
```

### Make regression report

OLS Regression Results Interpretation - [Click Here](https://www.geeksforgeeks.org/interpreting-the-results-of-linear-regression-using-ols-summary/)


```python
# create data for fitting
X_obs, X_sim = ds_obs, ds_sim

# add constant to predictor variables
X_obs = sm.add_constant(X_obs) 

# fit linear regression model
model = sm.OLS(X_sim, X_obs).fit()

# view model summary
print(model.summary())
```

### Multiple Regression

```python
# import data
multiple_df = pd.read_csv('data/multiple_regression_data.csv')

multiple_df
```

Multiple Linear Regression

```python
# create data for fitting
X_data, y = multiple_df.iloc[:,:2], multiple_df.iloc[:,2]

# add constant to predictor variables
X = sm.add_constant(X_data) 

# fit linear regression model
model = sm.OLS(y, X).fit()

# view model summary
print(model.summary())
```

Multiple Linear Regression with Logarithmic Transformation

```python
# create data for fitting
X_data, y = np.log10(multiple_df.iloc[:,:2]), np.log10(multiple_df.iloc[:,2])

# add constant to predictor variables
X = sm.add_constant(X_data) 

# fit linear regression model
model = sm.OLS(y, X).fit()

# view model summary
print(model.summary())
```

### Import module

```python
import matplotlib.pyplot as plt
```

### Recall data for plotting

```python
# Observation 
ds_obs = ds['obs']

# Simulation
ds_sim = ds['sim']
```

### Create empty plot

```python
# Create 2 subplots in a figure
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15,5))

plt.show()
```

### Add data to subplot 1

```python
# Plotting regression
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

# Add line graph to subplot 1
ax1.plot(ds_obs, color="#2980B9", lw=1.5, alpha=0.7, label='Observed data')
ax1.plot(ds_sim, color="black", lw=1.5, linestyle='--', alpha=0.7, label='Simulated data')

plt.show()
```

### Customize subplot 1

```python
# Plotting regression
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

# Add line graph to subplot 1
ax1.plot(ds_obs, color="#2980B9", lw=1.5, alpha=0.7, label='Observed data')
ax1.plot(ds_sim, color="black", lw=1.5, linestyle='--', alpha=0.7, label='Simulated data')

# Customization
ax1.set_ylim(100, 220)
ax1.set_title('Observed and simulated shoreline\n', fontsize=15)
ax1.set_ylabel('\n Distance from transect line (m)', fontsize=12)
ax1.set_xlabel('Number of transect points', fontsize=12)
ax1.legend(loc='lower left')


plt.show()
```

### Add data to subplot 2

```python
# Plotting regression
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

# Add line graph to subplot 1
ax1.plot(ds_obs, color="#2980B9", lw=1.5, alpha=0.7, label='Observed data')
ax1.plot(ds_sim, color="black", lw=1.5, linestyle='--', alpha=0.7, label='Simulated data')

# Customization
ax1.set_ylim(100, 220)
ax1.set_title('Observed and simulated shoreline\n', fontsize=15)
ax1.set_ylabel('\n Distance from transect line (m)', fontsize=12)
ax1.set_xlabel('Number of transect points', fontsize=12)
ax1.legend(loc='lower right')

# Add line graph to subplot 2
ax2.plot(ds_obs, ds_sim, ls="", marker="o", color='black', markersize=2)


plt.show()
```

### Calculate 1:1 reference line

```python
# Calculate 1:1 reference line
percs = np.linspace(0,100,21)
qn_a = np.percentile(ds_obs, percs)
qn_b = np.percentile(ds_sim, percs)
x = np.linspace(np.min((qn_a.min(),qn_b.min())), np.max((qn_a.max(),qn_b.max())))
```

### Add 1:1 line to subplot 2

```python
# Plotting regression
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

# Add line graph to subplot 1
ax1.plot(ds_obs, color="#2980B9", lw=1.5, alpha=0.7, label='Observed data')
ax1.plot(ds_sim, color="black", lw=1.5, linestyle='--', alpha=0.7, label='Simulated data')

# Customization
ax1.set_ylim(100, 220)
ax1.set_title('Observed and simulated shoreline\n', fontsize=15)
ax1.set_ylabel('\n Distance from transect line (m)', fontsize=12)
ax1.set_xlabel('Number of transect points', fontsize=12)
ax1.legend(loc='lower right')

# Add line graph to subplot 2
ax2.plot(ds_obs, ds_sim, ls="", marker="o", color='black', markersize=2)
ax2.plot(x, x, color="#900C3F", ls="-", lw=2.0)

plt.show()
```

### Customize subplot 2

```python
# Plotting regression
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

# Add line graph to subplot 1
ax1.plot(ds_obs, color="#2980B9", lw=1.5, alpha=0.7, label='Observed data')
ax1.plot(ds_sim, color="black", lw=1.5, linestyle='--', alpha=0.7, label='Simulated data')

# Customization
ax1.set_ylim(100, 220)
ax1.set_title('Observed and simulated shoreline\n', fontsize=15)
ax1.set_ylabel('\n Distance from transect line (m)', fontsize=12)
ax1.set_xlabel('Number of transect points', fontsize=12)
ax1.legend(loc='lower right')

# Add line graph to subplot 2
ax2.plot(ds_obs, ds_sim, ls="", marker="o", color='black',  alpha=0.5, markersize=2)
ax2.plot(x,x, color="#900C3F", ls="-", lw=2.0)

# Customization
ax2.set_ylim(100, 220)
ax2.set_title('Regression between manual and automated shoreline\n', fontsize=15)
ax2.set_ylabel('\n Distance from transect line (m) - Simulated', fontsize=12)
ax2.set_xlabel('Distance from transect line (m) - Observed', fontsize=12)
ax2.text(120, 200, '$R^2 = 0.98$')
ax2.text(120, 195, '$RMSE = 2.85$')

plt.tight_layout()
plt.show()
```
